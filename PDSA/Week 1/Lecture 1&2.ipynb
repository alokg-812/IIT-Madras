# Compute primes from 1 to 100

```python
def factors(n):
    factorList = []
    for i in range(1, n+1):
        if(n%i==0):
            factorList.append(i)
    return factorList
```

```python
def primes(n):
    return factors(n) == [1,n]
```

```python
def primes(n):
    return(len(factors(n)) == 2)
```

```python
primeList = []
cnt_of_primes = 0
for i in range(1, 101):
    if(primes(i)):
        primeList.append(i)
        cnt_of_primes+=1
        
print(primeList)
print(cnt_of_primes)
```
